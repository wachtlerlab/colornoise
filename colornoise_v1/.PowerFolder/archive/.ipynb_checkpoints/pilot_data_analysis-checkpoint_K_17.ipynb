## 2.1 PF fitting by Psychopy

Psychopy can fit psychophysics data to a cumulative Gaussian function and return estimated parameters (centre and standard deviation). The goodness of fit is represented by the sum of squares (ssq).

The threshold is estimated where the perfomance is 75% correct (i.e. same as estimated Gaussian mean values). 

The circle size in the figure below is scaled by the weights of trial numbers. 

```python
from exploredata import ExploreData
```

### 2.1.1 Design 1

### Subject: ysu, hue set 1, hue set 2, and hue set 1+2

```python
data_ysu = ExploreData('ysu', sel_par=['cn2x8_LL_b', 'cn2x8_LL_c'], sel_ses=['202009'])
data_ysu.pfplot()
data_ysu.threshplot(polar=True)
```

```python
data_ysu_new.paramplot()
```

#### Combine the two hue sets:

```python
ntrial_ysu_1, res_ysu_1 = data_ysu.fitpf()
ntrial_ysu_2, res_ysu_2 = data_ysu_new.fitpf()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
from exploredata import color4plot

N = 16

angles = np.repeat([(n / float(N) * 360) for n in range(N)], 2, axis=0)
color_codes = color4plot(N)

centre_1 = [res_ysu_1[k]['fit'].params[0] for k in res_ysu_1.keys()]
std_1 = [res_ysu_1[k]['fit'].params[1] for k in res_ysu_1.keys()]
par_err_1 = [np.sqrt(np.diagonal(res_ysu_1[k]['fit'].covar)) for k in res_ysu_1.keys()]

centre_2 = [res_ysu_2[k]['fit'].params[0] for k in res_ysu_2.keys()]
std_2 = [res_ysu_2[k]['fit'].params[1] for k in res_ysu_2.keys()]
par_err_2 = [np.sqrt(np.diagonal(res_ysu_2[k]['fit'].covar)) for k in res_ysu_2.keys()]


centre = list(itertools.chain(*zip(centre_1, centre_2)))
std = list(itertools.chain(*zip(std_1, std_2)))
par_err = list(itertools.chain(*zip(par_err_1, par_err_2)))


plt.figure(figsize=(16, 8))
ax = plt.subplot(111)
ax.set_title('cumulative Gaussian params')
ax.scatter(range(2 * N), centre, s=70, color=np.repeat(color_codes, repeats=2, axis=0), marker='o', label='mean')
ax.scatter(range(2 * N), std, s=70, color=np.repeat(color_codes, repeats=2, axis=0), marker='v', label='std')
ax.errorbar(range(2 * N), centre, yerr=[x[0] for x in par_err], label='mean', color='grey', ls='--', capsize=5)
ax.errorbar(range(2 * N), std, yerr=[x[1] for x in par_err], label='std', color='silver', ls='--', capsize=5)

# xlabels = [f"{l}\n{a}" for l, a in zip(labels, angles)]
ax.set_xticks(range(2 * N))
ax.set_xticklabels(angles)
plt.xticks(rotation=70)
plt.legend()
```
