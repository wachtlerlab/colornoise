## 2.1 PF fitting by Psychopy

```python
from exploredata import ExploreData

data_ysu = ExploreData('ysu', sel_ses=['202009'])
```

Psychopy can fit psychophysics data to a cumulative Gaussian function and return estimated parameters (centre and standard deviation). The goodness of fit is represented by the sum of squares (ssq).

The threshold is estimated where the perfomance is 75% correct (i.e. same as estimated Gaussian mean values). 

The circle size in the figure below is scaled by the weights of trial numbers. 

```python
data_ysu.pfplot()
```

```python
data_ysu.threshplot(savefig=True)
```

```python
data_ysu.threshplot(radar=True, savefig=True)
```

```python
data_ysu.paramplot()
```
