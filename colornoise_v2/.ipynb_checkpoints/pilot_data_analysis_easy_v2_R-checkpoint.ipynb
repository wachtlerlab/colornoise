## L-L data v.s. H-H data (with uniform or scaled noise levels)

```r
data_ysu_easy = fit_pf('ysu', sel_par='cn2x8_LL_easy_a.yaml', rm_ses='0917T10')  # 10 sessions after removing 2 sessions 
hh_data_1 = fit_pf('ysu', sel_par='cn2x8_HH_ysu_test.yaml', rm_ses='0930|1001T10') # 8 sessions after removing 2 sessions 
hh_data_2 = fit_pf('ysu', sel_par='cn2x8_HH_ysu_scaledN_a.yaml', rm_ses='20201012')
```

```r
param_ll = param_plot(data_ysu_easy, plot=FALSE)
param_hh_1 = param_plot(hh_data_1, plot=FALSE)
param_hh_2 = param_plot(hh_data_2, plot=FALSE)

p1_ll <- param_ll %>% filter(parn=='p1')
p2_ll <- param_ll %>% filter(parn=='p2')

p1_hh_1 <- param_hh_1 %>% filter(parn=='p1')
p2_hh_1 <- param_hh_1 %>% filter(parn=='p2')

p1_hh_2 <- param_hh_2 %>% filter(parn=='p1')
p2_hh_2 <- param_hh_2 %>% filter(parn=='p2')
```

```r
col <- c('LL', 'HH_uniformN', 'HH_scaledN')

plt2 <- ggplot() +
    geom_point(data=p2_ll, aes(x=angle, y=par, color=col[1])) +
    geom_line(data=p2_ll, aes(x=angle, y=par, color=col[1])) +
    geom_pointrange(data=p2_ll, aes(x=angle, y=par, ymin=parinf, ymax=parsup, color=col[1])) + 

    geom_point(data=p2_hh_1, aes(x=angle, y=par, color=col[2])) +
    geom_line(data=p2_hh_1, aes(x=angle, y=par, color=col[2])) +
    geom_pointrange(data=p2_hh_1, aes(x=angle, y=par, ymin=parinf, ymax=parsup, color=col[2])) + 

    geom_point(data=p2_hh_2, aes(x=angle, y=par, color=col[3])) +
    geom_line(data=p2_hh_2, aes(x=angle, y=par, color=col[3])) +
    geom_pointrange(data=p2_hh_2, aes(x=angle, y=par, ymin=parinf, ymax=parsup, color=col[3])) 
    
plt2 + ggtitle('JND, i.e. Gaussian std') +
  xlab("Hue angle (degree)") + ylab("Estimated parameter (degree)") +
    theme(text=element_text(size=20))

plt1 <- ggplot() +
    geom_point(data=p1_ll, aes(x=angle, y=par, color=col[1])) +
    geom_line(data=p1_ll, aes(x=angle, y=par, color=col[1])) +
    geom_pointrange(data=p1_ll, aes(x=angle, y=par, ymin=parinf, ymax=parsup, color=col[1])) + 

    geom_point(data=p1_hh_1, aes(x=angle, y=par, color=col[2])) +
    geom_line(data=p1_hh_1, aes(x=angle, y=par, color=col[2])) +
    geom_pointrange(data=p1_hh_1, aes(x=angle, y=par, ymin=parinf, ymax=parsup, color=col[2])) + 

    geom_point(data=p1_hh_2, aes(x=angle, y=par, color=col[3])) +
    geom_line(data=p1_hh_2, aes(x=angle, y=par, color=col[3])) +
    geom_pointrange(data=p1_hh_2, aes(x=angle, y=par, ymin=parinf, ymax=parsup, color=col[3])) 
    
plt1 + ggtitle("PSE, i.e. Gaussian mean") +
  xlab("Hue angle (degree)") + ylab("Estimated parameter (degree)")+
    theme(text=element_text(size=20))

figure <- ggarrange(plt2, plt1, ncol = 1, nrow = 2)
```
