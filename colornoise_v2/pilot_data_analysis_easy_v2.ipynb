## L-L data v.s. H-H data (with uniform or scaled noise levels)

```python
from exploredata import ExploreData
data_ysu_easy = ExploreData('ysu', sel_par=['cn2x8_LL_easy_a.yaml'], rm_ses=['0917T10'])  # 10 sessions after removing 2 sessions 
hh_data_1 = ExploreData('ysu', sel_par=['cn2x8_HH_ysu_test.yaml'], rm_ses=['0930','1001T10']) # 8 sessions after removing 2 sessions 
hh_data_2 = ExploreData('ysu', sel_par=['cn2x8_HH_ysu_scaledN_a.yaml'], rm_ses=['20201012']) # 9 sessions after removing 3 sessions
```

```python
hh_data_1.rearrange_pfplot()
```

```python
hh_data_2.rearrange_pfplot()
```

```python
import matplotlib.pyplot as plt
import numpy as np

rearr_ll = data_ysu_easy.rearrange()
rearr_hh_1 = hh_data_1.rearrange()
rearr_hh_2 = hh_data_2.rearrange()

num = len(rearr_ll)
angles = [(n / float(num) * 360 + 22.5) for n in range(num)]

centre_ll = rearr_ll['fit'].map(lambda x: x.params[0])
std_ll = rearr_ll['fit'].map(lambda x: x.params[1])
par_err_ll = rearr_ll['fit'].map(lambda x: np.sqrt(np.diagonal(x.covar)))

centre_hh_1 = rearr_hh_1['fit'].map(lambda x: x.params[0])
std_hh_1 = rearr_hh_1['fit'].map(lambda x: x.params[1])
par_err_hh_1 = rearr_hh_1['fit'].map(lambda x: np.sqrt(np.diagonal(x.covar)))

centre_hh_2 = rearr_hh_2['fit'].map(lambda x: x.params[0])
std_hh_2 = rearr_hh_2['fit'].map(lambda x: x.params[1])
par_err_hh_2 = rearr_hh_2['fit'].map(lambda x: np.sqrt(np.diagonal(x.covar)))

plt.subplots(2, 1, figsize=(12, 10))
colorcodes = ['limegreen', 'skyblue', 'steelblue']

ax1 = plt.subplot(211)
ax1.set_title('JND, i.e. Gaussian std')

ax1.errorbar(range(num), std_ll, yerr=[x[1] for x in par_err_ll], color=colorcodes[0], ls='-')
ax1.scatter(range(num), std_ll, s=80, color=colorcodes[0], label='std/JND/threshold, LL', marker='v')

ax1.errorbar(range(num), std_hh_1, yerr=[x[1] for x in par_err_hh_1], color=colorcodes[1], ls='-')
ax1.scatter(range(num), std_hh_1, s=80, color=colorcodes[1], label='std/JND/threshold, HH_uniformN', marker='v')

ax1.errorbar(range(num), std_hh_2, yerr=[x[1] for x in par_err_hh_2], color=colorcodes[2], ls='-')
ax1.scatter(range(num), std_hh_2, s=80, color=colorcodes[2], label='std/JND/threshold, HH_scaledN', marker='v')
ax1.set_ylim([0, 16])
ax1.legend()

ax1.set_xticks(range(num))
ax1.set_xticklabels(angles)
plt.xticks(rotation=45)
ax1.set_ylabel('hue angle (degree)')

ax2 = plt.subplot(212)
ax2.set_title('PSE, i.e. Gaussian mean')

ax2.errorbar(range(num), centre_ll, yerr=[x[0] for x in par_err_ll], color=colorcodes[0], ls='--')
ax2.scatter(range(num), centre_ll, s=80, color=colorcodes[0], label='mean/PSE, LL', marker='o')

ax2.errorbar(range(num), centre_hh_1, yerr=[x[0] for x in par_err_hh_1], color=colorcodes[1], ls='--')
ax2.scatter(range(num), centre_hh_1, s=80, color=colorcodes[1], label='mean/PSE, HH_uniformN', marker='o')

ax2.errorbar(range(num), centre_hh_2, yerr=[x[0] for x in par_err_hh_2], color=colorcodes[2], ls='--')
ax2.scatter(range(num), centre_hh_2, s=80, color=colorcodes[2], label='mean/PSE, HH_scaledN', marker='o')

ax2.plot(range(num), np.repeat(0, num), '--', color='lightgrey')

ax2.legend()

ax2.set_xticks(range(num))
ax2.set_xticklabels(angles)
plt.xticks(rotation=45)
ax2.set_xlabel('hue')
ax2.set_ylabel('hue angle (degree)')

plt.show()
```

## Compare external noise effects: uniform noise level v.s. scaled noise level

```python
import matplotlib.pyplot as plt
import numpy as np

rearr_ll = data_ysu_easy.rearrange()
rearr_hh_1 = hh_data_1.rearrange()
rearr_hh_2 = hh_data_2.rearrange()

num = len(rearr_ll)
angles = [(n / float(num) * 360 + 22.5) for n in range(num)]

centre_ll = rearr_ll['fit'].map(lambda x: x.params[0])
std_ll = rearr_ll['fit'].map(lambda x: x.params[1])
par_err_ll = rearr_ll['fit'].map(lambda x: np.sqrt(np.diagonal(x.covar)))

centre_hh_1 = rearr_hh_1['fit'].map(lambda x: x.params[0])
std_hh_1 = rearr_hh_1['fit'].map(lambda x: x.params[1])
par_err_hh_1 = rearr_hh_1['fit'].map(lambda x: np.sqrt(np.diagonal(x.covar)))

centre_hh_2 = rearr_hh_2['fit'].map(lambda x: x.params[0])
std_hh_2 = rearr_hh_2['fit'].map(lambda x: x.params[1])
par_err_hh_2 = rearr_hh_2['fit'].map(lambda x: np.sqrt(np.diagonal(x.covar)))

plt.subplots(1, 2, figsize=(18, 10))
colorcodes = ['limegreen', 'skyblue', 'steelblue']

ax1 = plt.subplot(121)
ax1.set_title('H-H with uniform noise levels')

ax1.errorbar(range(num), std_ll, yerr=[x[1] for x in par_err_ll], color=colorcodes[0], ls='-')
ax1.scatter(range(num), std_ll, s=80, color=colorcodes[0], label='std/JND/threshold, LL', marker='v')

ax1.errorbar(range(num), centre_ll, yerr=[x[0] for x in par_err_ll], color=colorcodes[0], ls='--')
ax1.scatter(range(num), centre_ll, s=80, color=colorcodes[0], label='mean/PSE, LL', marker='o')

ax1.errorbar(range(num), std_hh_1, yerr=[x[1] for x in par_err_hh_1], color=colorcodes[1], ls='-')
ax1.scatter(range(num), std_hh_1, s=80, color=colorcodes[1], label='std/JND/threshold, HH_uniformN', marker='v')

ax1.errorbar(range(num), centre_hh_1, yerr=[x[0] for x in par_err_hh_1], color=colorcodes[1], ls='--')
ax1.scatter(range(num), centre_hh_1, s=80, color=colorcodes[1], label='mean/PSE, HH_uniformN', marker='o')

ax1.plot(range(num), np.repeat(0, num), '--', color='lightgrey')
ax1.legend()

ax1.set_xticks(range(num))
ax1.set_xticklabels(angles)
plt.xticks(rotation=70)
ax1.set_xlabel('hue')
ax1.set_ylabel('params')

ax2 = plt.subplot(122)
ax2.set_title('H-H with scaled noise levels')

ax2.errorbar(range(num), std_ll, yerr=[x[1] for x in par_err_ll], color=colorcodes[0], ls='-')
ax2.scatter(range(num), std_ll, s=80, color=colorcodes[0], label='std/JND/threshold, LL', marker='v')

ax2.errorbar(range(num), centre_ll, yerr=[x[0] for x in par_err_ll], color=colorcodes[0], ls='--')
ax2.scatter(range(num), centre_ll, s=80, color=colorcodes[0], label='mean/PSE, LL', marker='o')

ax2.errorbar(range(num), std_hh_2, yerr=[x[1] for x in par_err_hh_2], color=colorcodes[2], ls='-')
ax2.scatter(range(num), std_hh_2, s=80, color=colorcodes[2], label='std/JND/threshold, HH_scaledN', marker='v')

ax2.errorbar(range(num), centre_hh_2, yerr=[x[0] for x in par_err_hh_2], color=colorcodes[2], ls='--')
ax2.scatter(range(num), centre_hh_2, s=80, color=colorcodes[2], label='mean/PSE, HH_scaledN', marker='o')

ax2.plot(range(num), np.repeat(0, num), '--', color='lightgrey')

ax2.legend()

ax2.set_xticks(range(num))
ax2.set_xticklabels(angles)
plt.xticks(rotation=70)
ax2.set_xlabel('hue')
ax2.set_ylabel('params')

plt.show()
```
